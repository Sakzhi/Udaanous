```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('web_traffic.csv')
df.head()
```

```python
df['Date']= pd.to_datetime(df['Date'])
df.dtypes
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='SessionDuration',y='Device',data=df)
plt.title('Average session time')
plt.xlabel('SessionDuration')
plt.ylabel('Device')
plt.show()
```

```python
plt.figure(figsize=(6,4)) 
sns.barplot(x ='UserType',y='PagesVisited', data = df)
plt.title('Pages visited by usertype')  
plt.xlabel('User Type')  
plt.ylabel('Pages Visited')  
plt.show()  
```

```python
sessiontime=df.groupby('Date')['SessionDuration'].sum()
sessiontime.plot(kind='bar', title='Trend Over Time', figsize=(9,5))
plt.show()
```

```python
a= df.groupby('Device')['SessionDuration'].sum()
a.plot(kind='pie',autopct='%1.2f%%',title='Device usage distribution', figsize=(10,5))
plt.show()
```

```python
b = df.groupby('Device')['PagesVisited'].sum()
b.plot(kind='bar', title='Average pages visited', figsize=(10,5))
plt.show()
```

```python
peak_days = df.sort_values(by='VisitID', ascending=False)
print(peak_days.head())  # Shows top traffic days
```

```python
 plt.figure(figsize=(10,6))
 sns.lineplot(x='VisitID', y='Date', data=df, marker='o')
 plt.title('')
 plt.xlabel(' ')
 plt.ylabel('')
 plt.show()
```
