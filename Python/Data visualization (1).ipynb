```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a real-world dataset for correlation analysis
data = pd.DataFrame({
    'customer_id': range(101, 111),
    'electronics_sales': [2500, 1200, 3400, 1500, 2200, 1700, 2800, 3100, 4000, 2300],
    'clothing_sales': [1800, 900, 2100, 1200, 1750, 1100, 1900, 2400, 3100, 1800],
    'furniture_sales': [3000, 1400, 2700, 1600, 2500, 1900, 3100, 3500, 4200, 2800],
    'grocery_sales': [900, 600, 800, 500, 700, 750, 850, 1100, 1300, 900]
})

# Calculating the correlation matrix
correlation_matrix = data.drop(columns=['customer_id']).corr()

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Sales Across Product Categories')
plt.show()
```
