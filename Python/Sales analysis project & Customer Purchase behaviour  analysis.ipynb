```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df = pd.read_excel('practice_data.xlsx')
df.head()
```

```python
x=df[['Interpersonal Skills Rating']]
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(x) 
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()    
```

```python
from sklearn.cluster import KMeans
import numpy as np

# Sample customer spending data
X = np.array([[200, 50], [180, 60], [20, 250], [30, 230], [190, 55]])

# Define number of clusters (let's say 2 groups)
selected_k_value = 2
kmeans = KMeans(n_clusters=selected_k_value, random_state=42)

# Fit model
kmeans.fit(X)

# Get cluster labels
labels = kmeans.labels_
print(labels)  # Output might be something like [0, 0, 1, 1, 0]
```
