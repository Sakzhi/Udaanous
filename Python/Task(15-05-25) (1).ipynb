# datetime function

```python
import pandas as pd

# Sample dataset with a date column
data2 = pd.DataFrame({
    "date_variable": pd.to_datetime(["2024-05-15", "2023-07-22", "2022-11-10", "2025-01-30", "2021-09-05"])
})

# Extracting year, month, and day
data2["year"] = data2["date_variable"].dt.year
data2["month"] = data2["date_variable"].dt.month
data2["day"] = data2["date_variable"].dt.day

# Display the dataframe
print(data2)
```

# feature emcoding

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Sample dataset
data3 = pd.DataFrame({
    "Customer_Type": ["Regular", "Premium", "Regular", "VIP", "Premium"]
})

# Initialize LabelEncoder
encoder = LabelEncoder()

# Fit and transform
data3["Encoded_Type"] = encoder.fit_transform(data3["Customer_Type"])

# Display results
print(data3)
```

# feature mapping

```python
data4=pd.DataFrame({'job role':['data analyst','Manager','Engineer','data science']})
mapping={'data analyst' : 1,'Manager': 2,'Engineer':3,'data science':4}
data4['job role mapped']=data4['job role'].map(mapping)
print(data4)
```

# feature binning

```python
import pandas as pd

# Sample dataset
data5 = pd.DataFrame({'purchase_amount': [150, 800, 3000, 120, 450, 2500, 900]})

# Define bins and labels
bins = [0, 500, 2000, 5000]  # Ranges for Low, Medium, High spending
labels = ['Low', 'Medium', 'High']

# Apply binning
data5['spending_category'] = pd.cut(data5['purchase_amount'], bins=bins, labels=labels, include_lowest=True)

print(data5)
```
