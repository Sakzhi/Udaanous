```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df = pd.read_excel('practice_data.xlsx')
df.head()
```

```python
x=df[['Interpersonal Skills Rating']]
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(x) 
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()    
```

```python
import pandas as pd
import numpy as np
data1= pd.read_csv('sales.csv')
print(data1.head())
```

```python
missing_data = data1.isnull().sum()
print("Missing Data")
missing_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# barchart 

```python
plt.figure(figsize=(6,4)) 
sns.barplot(x = 'Salesperson',y='Sales', data = data1)
plt.title('Sales performance')  
plt.xlabel('Salesperson')  
plt.ylabel('Sales')  
plt.show()  
```

# hist plot

```python
 plt.figure(figsize=(8,8))
 sns.histplot(data1['Sales'], bins=5)
 plt.title('Sales performance')
 plt.show()
```

# scatter plot

```python
 plt.figure(figsize=(6,6))
 plt.scatter(data1['Sales'], data1['OrderID'])
 plt.title('Sales Performance')
 plt.xlabel('Sales')
 plt.ylabel('OrderID')
 plt.show()
```

# box plot

```python
 plt.figure(figsize=(6,6))
 sns.boxplot(x='Sales', y='ProductName',data=data1)
 plt.title('Sales Performance')
 plt.xlabel('Sales')
 plt.ylabel('ProductName')
 plt.show()
```

# pie chart

```python
plt.figure(figsize=(6, 6))
plt.pie(data1['Sales'],autopct='%1.2f%%',colors=['blue', 'green', 'red', 'purple'])
plt.title('Sales Performance') 
plt.ylabel(' ')
plt.show()
```

# line plot

```python
 plt.figure(figsize=(5,5))
 sns.lineplot(x='Region', y='Sales', data=data1, marker='o')
 plt.title('Region wise-sales')
 plt.xlabel('Region')
 plt.ylabel('Sales')
 plt.show()
```

# bar plot

```python
data1.set_index("OrderDate").plot(kind="bar", stacked=True)
plt.title("Stacked Sales Data")
plt.xlabel("Order Date")
plt.ylabel("Sales")
plt.legend(title="Salesperson")
plt.show()
```
